### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2

```python
heads = 10
legs = 28

# Using equations:
# x + y = heads
# 4x + 2y = legs

# Solve:
x = (legs - 2 * heads) // 2  # Dogs
y = heads - x                # Chickens

print("Dogs: ", x)
print("Chickens: ", y)


```
